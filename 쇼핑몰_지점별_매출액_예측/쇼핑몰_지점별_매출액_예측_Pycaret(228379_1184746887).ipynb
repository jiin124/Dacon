# 데이터 나누기

```python
def seprate_data(concat_Alldata):
    train = concat_Alldata[pd.notnull(concat_Alldata['Weekly_Sales'])]
    test = concat_Alldata[~pd.notnull(concat_Alldata['Weekly_Sales'])]
    return train, test

train, test = seprate_data(all_data)
```

```python
train.head()
```

```python
test.head()
```

```python
train['Weekly_Sales']=np.log1p(train['Weekly_Sales'])
```

```python
print(train.shape)
print(test.shape)
```

```python
from pycaret.regression import *
```

```python
model = setup(train, target = 'Weekly_Sales', 
              fold_shuffle=True,
              fold=15, 
              session_id=530,
              remove_multicollinearity = True,
              normalize= True,
              n_jobs = -1,
              silent = True,
              combine_rare_levels = True,
              ignore_low_variance = True )
```

```python
top5_model = compare_models(round=4, n_select = 5, sort='RMSE')
```

```python
stacker = stack_models(top5_model[:4], optimize = 'RMSE')
```

```python
prediction = predict_model(stacker, data=test)
pred = np.expm1(prediction['Label'])

temp = pd.read_csv('sample_submission.csv')
temp['Weekly_Sales'] = pred.reset_index()['Label']

submission = pd.DataFrame({
        "id" : temp['id'],
        "Weekly_Sales" : temp['Weekly_Sales']
})
submission.to_csv('submit1.csv', index = False)
```
